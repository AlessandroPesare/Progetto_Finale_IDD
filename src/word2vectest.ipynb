```python
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
```

```python
# Esempio di record di testo
record1 = "Questo è il primo record di testo"
record2 = "Questo è il secondo record di testo"
record3 = "Ciao belli sono diverso dagli altri"
record4 = "Stiamo lavorando con embedding di string"
```

```python
# Creazione di una lista di record
records = [record1.split(), record2.split(), record3.split(),record4.split()]
```

```python
print(records)
```

```python
model = Word2Vec(sentences=records, vector_size=100, window=5, min_count=1, workers=4)
```

```python
vector_record1 = model.wv[record1.split()]
vector_record2 = model.wv[record2.split()]
vector_record3 = model.wv[record3.split()]
vector_record4 = model.wv[record4.split()]
```

```python
print(vector_record1)
```

```python
def calculate_vector(record):
    # Inizializza un vettore vuoto
    record_vector = np.zeros(model.vector_size)
    # Somma i vettori di embedding di ciascun token nel record
    for token in record:
        record_vector += model.wv[token]
    # Normalizza il vettore risultante
    record_vector /= 100
    return record_vector
```

```python
vector_record_1_embedded = calculate_vector(record1.split())
vector_record2_embedded = calculate_vector(record2.split())
vector_record3_embedded = calculate_vector(record3.split())
vector_record4_embedded = calculate_vector(record4.split())
```

```python
print("Vettore record 1:", vector_record_1_embedded)
print("Vettore record 2:", vector_record2_embedded)
print("Vettore record 3:", vector_record3_embedded)
print("Vettore record 4:", vector_record4_embedded)
```

```python
# Combinazione di vettori in un'unica matrice
vectors = [vector_record_1_embedded, vector_record2_embedded, vector_record3_embedded,vector_record4_embedded]
```

## CALCOLO DISTRIBUZIONE DEI CLUSTER GENERATI DAL BLOCKING


##### CREO IL FILE TXT --> CLUSTERi:QUANTI RECORDS HA

```python
# dizionario per mantenere le occorrenze di ciascun cluster
cluster_to_record = {}

with open('./../clustering_full_24.txt', 'r') as file:
    for line in file:
        # divido la riga in due parti usando il ':'
        parts = line.strip().split(':')
        if len(parts) == 2:
            # ottengo il cluster corrente 
            cluster_corrente = parts[1].strip()
            cluster_to_record[cluster_corrente] = cluster_to_record.get(cluster_corrente, 0) + 1


for cluster, records in cluster_to_record.items():
    print(f"Il cluster {cluster} ha {records} record.")
```

```python
with open('stats_blocking.txt', 'w') as outfile:
    for cluster, records in cluster_to_record.items():
        outfile.write(f"{cluster}:{records}\n")
```

##### CREO IL FILE TXT --> NUM RECORD : QUANTI CLUSTER HANNO QUELLA DIMENSIONE

```python
num_record_to_cluster = {}

with open('stats_blocking_GPT.txt', 'r') as file:
    for line in file:
        parts = line.strip().split(':')
        if len(parts) == 2:
            dim_cluster = parts[1].strip()
            num_record_to_cluster[dim_cluster] = num_record_to_cluster.get(dim_cluster, 0) + 1


for dim_cluster, num_cluster in num_record_to_cluster.items():
    print(f"Di dimensione {dim_cluster} ci sono {num_cluster} cluster.")
```

```python
# converto il dizionario in una lista di tuple per ordinarlo in base al cluster
num_record_to_cluster_sorted = sorted(num_record_to_cluster.items(), key=lambda x: int(x[0]))

with open('stats_blockingGPT_frequencyClusterDimension.txt', 'w') as outfile:
    for dim, cluster in num_record_to_cluster_sorted:
        outfile.write(f"{dim}:{cluster}\n")
```

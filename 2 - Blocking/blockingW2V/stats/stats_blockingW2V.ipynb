#### CALCOLO DISTRIBUZIONE DEI CLUSTER GENERATI DAL BLOCKING


```python
import csv

# dizionario per mantenere le occorrenze di ciascun cluster
cluster_to_record = {}

with open('SchemaMediatoConCluster.csv', newline='') as csvfile:
    csv_reader = csv.DictReader(csvfile)
    for row in csv_reader:
        cluster = row['block']
        cluster_to_record[cluster] = cluster_to_record.get(cluster, 0) + 1

for cluster, records in cluster_to_record.items():
    print(f"Il cluster {cluster} ha {records} record.")
```

```python
# converto il dizionario in una lista di tuple per ordinarlo in base al cluster
cluster_to_record_sorted = sorted(cluster_to_record.items(), key=lambda x: int(x[0]))

with open('stats_blockingW2V.txt', 'w') as outfile:
    for cluster, records in cluster_to_record_sorted:
        outfile.write(f"{cluster}:{records}\n")
```
